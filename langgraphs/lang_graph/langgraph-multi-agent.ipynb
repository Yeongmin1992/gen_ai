### HandOffs (핸드오프)
- 한 에이전트가 다른 에어전트에게 제어권을 넘겨주는 Multi Agent 상호작용 방식

- Command 사용하기

```python
from typing import Literal
from langgraph.types import Command
from langgraph.graph import StateGraph, MessagesState, START, END
from lnagchain_core.messages import AIMessage

def agent_1(state: MessageState):
    messages = AIMessage(content="This is a message from agent 1.")
    # 어떤 에이전트 노드로 갈지 지정 및 상태 업데이트 가능
    return Command(goto="agent_2", updates={"messages": [messages]})

def agent_2(state: MessageState):
    messages = AIMessage(content="This is a message from agent 2.")
    return Command(goto=END, updates={"messages": [messages]})

graph_builder = StateGraph(MessagesState)
# 위에서 어느 노드로 갈지 지정해주어 agent_1에서 agent_2로 이동하는 엣지 추가하지 않아도 됨
graph_builder.add_node("agent_1", agent_1)
graph_builder.add_node("agent_2", agent_2)

graph = graph_builder.build(START, "agent_1")

graph = graph_builder.compile()
```
